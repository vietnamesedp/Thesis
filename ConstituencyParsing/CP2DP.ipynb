```python
class Word:

  def __init__(self, id = '-', form = '-', lemma = '-', upos = '-', xpos = '-', feats = '-', head = '-', deprel = '-', deps = '-', misc = '-'):
    self.id = id  # ID của từ
    self.form = form  # Hình thức bề mặt của từ
    self.lemma = lemma  # Lemma của từ
    self.upos = upos  # Thẻ UPOS (Universal Part-of-Speech)
    self.xpos = xpos  # Thẻ XPOS (Extended Part-of-Speech)
    self.feats = feats  # Các thuộc tính hình thái của từ
    self.head = head  # ID của từ chủ ngữ
    self.deprel = deprel  # Mối quan hệ phụ thuộc của từ
    self.deps = deps  # Danh sách các từ phụ thuộc
    self.misc = misc  # Thông tin khác về từ

  def __str__(self):
    """Trả về chuỗi biểu diễn đối tượng Word."""
    return f"{self.id}\t{self.form}\t{self.lemma}\t{self.upos}\t{self.xpos}\t{self.feats}\t{self.head}\t{self.deprel}\t{self.deps}\t{self.misc}"
```

```python
import re

def get_pair(sentence):
  matches = re.findall(r'\(([\w\:\-]+) (\S+)\)', sentence)

  # Tìm tất cả các cặp label và từ tương ứng
  matches = re.findall(r'\(([\w\:\-]+) (\S+)\)', sentence)

  # Biến lưu trữ các cặp label và từ sau khi được xử lý
  processed_pairs = []

  # Biến lưu trữ từ tạm thời của nhãn "NNP"
  temp_word = ""

  # Duyệt qua danh sách các cặp label và từ
  for i, (label, word) in enumerate(matches, start = 1):
      if label == "NNP":  # Nếu nhãn là "NNP"
          temp_word += " " + word  # Nối từ vào từ tạm thời
      else:
          if temp_word:  # Nếu có từ tạm thời
              processed_pairs.append(("NNP", temp_word.strip().replace(")","")))  # Thêm từ tạm thời vào danh sách đã xử lý
              temp_word = ""  # Reset từ tạm thời
          if word.replace(")","") not in ["*E*", "*T*"]:
            processed_pairs.append((label, word.replace(")","")))  # Thêm cặp label và từ vào danh sách đã xử lý

  # Xử lí nếu từ tạm thời vẫn còn
  if temp_word:
      processed_pairs.append(("NNP", temp_word.strip()))



  return processed_pairs
```

```python
def get_parent_label(subtree):
  return str(subtree.parent()).split()[0].replace('(','')

def get_left_label(subtree):
  return str(subtree.left_sibling()).split()[0].replace('(','')

def get_right_label(subtree):
  return str(subtree.right_sibling()).split()[0].replace('(','')

def check_starts_with_label(word, label_list):
    return any(word.startswith(label) for label in label_list)

def is_form_label_word(tree):
  return bool(re.match(r'\(([\w\:\-]+) (\S+)\)', str(tree)))
```

```python
def get_root_vp_subtree(list_vp_subtree, pairs):
  if len(list_vp_subtree) >= 1:
    for subtree in list_vp_subtree:
      for s in subtree.subtrees(lambda t: t.height() == 2):
          pair = str(s).replace('(','').replace(')','').split()
          if pair[0].upper() == 'V-H' and pair[1] not in ["*T*","*E*", "*0*"]:
            root_word = pair[1]


          # root_pair = [pair for pair in pairs if pair[0].upper() == 'V-H']
            id = pairs.index((pair[0], pair[1])) + 1
            form = pair[1]
            lemma = form.lower()
            upos = pair[0]
            xpos = upos
            if lemma == 'là':
              return None
            root = Word(id = id, form = form, lemma = lemma, upos = upos, xpos = xpos, head = 0, deprel = "root")

            return root

def get_root_ap_subtree(list_ap_subtree, pairs):
  if len(list_ap_subtree) >= 1:
    for subtree in list_ap_subtree:
      for s in subtree.subtrees(lambda t: t.height() == 2):
          pair = str(s).replace('(','').replace(')','').split()
          if pair[0].upper() == 'ADJ-H' and pair[1] not in ["*T*","*E*", "*0*"]:
            root_word = pair[1]


          # root_pair = [pair for pair in pairs if pair[0].upper() == 'V-H']
            id = pairs.index((pair[0], pair[1])) + 1
            form = pair[1]
            lemma = form.lower()
            upos = pair[0]
            xpos = upos

            root = Word(id = id, form = form, lemma = lemma, upos = upos, xpos = xpos, head = 0, deprel = "root")

            return root

def get_root_np_subtree(list_np_subtree, pairs):
  if len(list_np_subtree) >= 1:
    for subtree in list_np_subtree:
      for s in subtree.subtrees(lambda t: t.height() == 2):
          pair = str(s).replace('(','').replace(')','').split()
          if pair[0].upper() == 'N-H' and pair[1] not in ["*T*","*E*", "*0*"]:
            root_word = pair[1]


          # root_pair = [pair for pair in pairs if pair[0].upper() == 'V-H']
            id = pairs.index((pair[0], pair[1])) + 1
            form = pair[1]
            lemma = form.lower()
            upos = pair[0]
            xpos = upos

            root = Word(id = id, form = form, lemma = lemma, upos = upos, xpos = xpos, head = 0, deprel = "root")

            return root

def get_root(cp_tree, pairs):
  # print(pairs)
  vp_cluster_1 = []
  vp_cluster_2 = []
  s_cluster = []
  ap_cluster = []
  np_cluster = []

  for subtree in cp_tree.subtrees():
    cluster_label = str(subtree).split()[0].replace("(", "")
    if cluster_label in cluster_labels and cluster_label.startswith("VP"):
      if subtree.parent() == cp_tree:
        vp_cluster_1.append(subtree)
      elif subtree.parent().parent() == cp_tree:
        vp_cluster_2.append(subtree)
    if cluster_label.startswith("S") and subtree != cp_tree:
      s_cluster.append(subtree)
    if cluster_label.startswith("AP"):
      if subtree.parent() == cp_tree:
        ap_cluster.append(subtree)
    if cluster_label.startswith("NP"):
      if subtree.parent() == cp_tree:
        np_cluster.append(subtree)

  # for x in s_cluster:
    # print(x)
  root = get_root_vp_subtree(vp_cluster_1, pairs)

  if root == None:
    for t in s_cluster:
      root = get_root(t, pairs)
      if root != None:
        return root

  if root == None:
    root = get_root_ap_subtree(ap_cluster, pairs)

  if root == None:
    root = get_root_np_subtree(np_cluster, pairs)

  if root == None:
     for t in vp_cluster_1:
      root = get_root(t, pairs)
      if root != None:
        return root

  # if cluster_label in cluster_labels and cluster_label.startswith("VP")


  return root
```

```python
 def get_previous_label_id(id, label, pairs):
  # print(id)
  if id is None:
    return -1

  for i in range(id - 2, -1, -1):
    if label.upper() == "N":
      if pairs[i][0].upper().startswith(label) and pairs[i][0].upper() not in  ["NUM","NUM-H","NC", "NC-H"]:
        return i + 1
      else:
        continue
    if pairs[i][0].upper().startswith(label):
      return i + 1

def get_previous_deprel_id(id, deprel, conllu_line):
  # print(id)
  if id is None:
    return -1

  for i in range(id - 2, -1, -1):
    if conllu_line[i].deprel == deprel:
      return i+1

def get_next_label_id(id, label, pairs):
  for i in range(id, len(pairs)):
    if label.upper() == "N":
      if pairs[i][0].upper().startswith(label) and pairs[i][0].upper() not in  ["NUM","NUM-H","NC", "NC-H"]:
        return i + 1
      else:
        continue
    if label.upper() == "-H":
      if pairs[i][0].upper().endswith(label) and pairs[i][0].upper() not in  ["NUM-H","NC-H"]:
        return i + 1
      else:
        continue
    if pairs[i][0].upper().startswith(label):
      return i+1
```

```python

import networkx as nx
import itertools

def is_subcycle(graph):
    """Tìm chu trình con trong đồ thị"""
    nodes = list(graph.nodes())
    for length in range(3, len(nodes) + 1):  # Chu trình tối thiểu phải có 3 đỉnh
        for subset in itertools.combinations(nodes, length):
            # Tạo đồ thị con với các đỉnh trong subset
            subgraph = graph.subgraph(subset)
            for cycle in nx.simple_cycles(nx.DiGraph(subgraph)):
                # Kiểm tra xem chu trình có phải là chu trình con không
                if len(cycle) > 2:
                    return True
    return False

def add_or_update_edge(graph, u, v):
    """Thêm hoặc cập nhật cạnh trong đồ thị có hướng.
    Nếu có một cạnh khác từ đỉnh u, cập nhật cạnh này với (u, v).
    """
    if u is None or v is None or v == '-':
      return

    if not graph.has_node(u):
        graph.add_node(u)

    if not graph.has_node(v):
        graph.add_node(v)

    # Kiểm tra nếu đỉnh u đã có cạnh ra
    outgoing_edges = list(graph.successors(u))

    if outgoing_edges:
        for target in outgoing_edges:
            if graph.has_edge(u, target):
                graph.remove_edge(u, target)

    graph.add_edge(u, v)


```

```python
def count_sentences(file_path):
  """
  This function will count the number of sentences in a file where sentences are separated by blank lines.

  Args:
    file_path: The path to the text file.

  Returns:
    The number of sentences in the file.
  """

  with open(file_path, 'r') as file:
    text = file.read()

  # Split the text into sentences based on blank lines
  sentences = text.split('\n\n')

  # Filter out any empty sentences that might result from extra blank lines
  sentences = [sentence for sentence in sentences if sentence.strip()]

  return len(sentences)

# Example usage with the provided file:
file_path = './content/system_file.txt'
num_sentences = count_sentences(file_path)
print(f"The number of sentences in '{file_path}' is: {num_sentences}")
```

```python
def fix_column1(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    with open(output_file, 'w', encoding='utf-8') as f:
        for line in lines:
            if line.strip():
                columns = line.strip().split('\t')

                if (columns[0] == '1'):
                  columns[1] = columns[1].lower()
                  columns[2] = columns[2].lower()

                columns[1] = columns[1].replace("_", " ")
                columns[2] = columns[2].replace("_", " ")
                f.write('\t'.join(columns) + '\n')
            else:
                f.write('\n')

input_file = './content/system_file.txt'
output_file = './content/output_fix_column1.txt'
fix_column1(input_file, output_file)

# input_file = '/content/single_gold_file.txt'
# output_file = 'output_gold.txt'
# fix_column1(input_file, output_file)
```

```python
from collections import Counter

def count_labels_from_conllu(file_path):
    # Bộ đếm cho các nhãn (cột thứ 8)
    label_counter = Counter()

    # Đọc file CoNLL-U
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if line and not line.startswith('#'):
                columns = line.split('\t')
                if len(columns) > 7:
                    label = columns[7]
                    label_counter[label] += 1

    # Trả về kết quả
    return label_counter

# Đường dẫn tới file CoNLL-U
file_path = './content/b_x.txt'

# Đếm số lần xuất hiện của các nhãn
label_counts = count_labels_from_conllu(file_path)

# Sắp xếp các nhãn theo thứ tự giảm dần dựa trên số lần xuất hiện
sorted_label_counts = sorted(label_counts.items(), key=lambda item: item[1], reverse=True)
sumtt = 0
# Hiển thị kết quả
for label, count in sorted_label_counts:
    print(f"{label}: {count}")
    sumtt += count

print(sumtt)
```

```python
def read_conllu(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = file.read().strip().split("\n\n")
    return [sentence.split("\n") for sentence in data]

def write_conllu(file_path, sentences):
    with open(file_path, 'w', encoding='utf-8') as file:
        for sentence in sentences:
            file.write("\n".join(sentence) + "\n\n")

def get_column_content(sentence, column_index):
    return "\n".join([line.split("\t")[column_index] for line in sentence if not line.startswith("#") and line.strip()])

def filter_sentences(file1_path, file2_path, output_path):
    # Đọc dữ liệu từ file1 và file2
    sentences_file1 = read_conllu(file1_path)
    sentences_file2 = read_conllu(file2_path)

    # Tạo tập hợp các nội dung cột thứ 2 từ file2
    column_content_set_file2 = {get_column_content(sentence, 1) for sentence in sentences_file2}
    print(len(column_content_set_file2))
    # Lọc các câu trong file1 mà cột thứ 2 của chúng có trong file2
    filtered_sentences = [sentence for sentence in sentences_file1 if get_column_content(sentence, 1) in column_content_set_file2]

    # Ghi các câu đã lọc vào file mới
    write_conllu(output_path, filtered_sentences)

# Đường dẫn tới các file
file1_path = './content/a_x.txt'
file2_path = './content/output_lowercase.txt'
output_path = './content/output_59xxx.txt'

# Thực hiện lọc và ghi kết quả vào file mới
filter_sentences(file1_path, file2_path, output_path)

# file3_path = 'output_lowercase.txt'
# file4_path = 'output_final.txt'
# output_path2 = 'output_final2.txt'

# # Thực hiện lọc và ghi kết quả vào file mới
# filter_sentences(file3_path, file4_path, output_path2)
```

```python
def extract_roots(sentences):
    roots = []
    for sentence in sentences:
        for line in sentence:
            if line.strip() and not line.startswith("#"):
                columns = line.split("\t")
                if columns[6] == "0":  # Cột thứ 8 là cột index 6 (0-based)
                    roots.append(columns[0])  # Cột thứ 2 là cột index 1 (0-based)
                    break
    return roots

def filter_sentences_by_common_roots(file1_path, file2_path, output1_path, output2_path):
    # Đọc dữ liệu từ file1 và file2
    sentences_file1 = read_conllu(file1_path)
    sentences_file2 = read_conllu(file2_path)

    # Extract roots from both files
    roots_file1 = extract_roots(sentences_file1)
    roots_file2 = extract_roots(sentences_file2)

    # Tìm các root chung
    # common_roots = set(roots_file1).intersection(set(roots_file2))
    # print(common_roots)

    # Lọc các câu có root trong danh sách common_roots
    filtered_sentences_file1 = []
    filtered_sentences_file2 = []

    for i in range(len(roots_file1)):
        if roots_file1[i] == roots_file2[i]:
          filtered_sentences_file1.append(sentences_file1[i])
          filtered_sentences_file2.append(sentences_file2[i])

    # Ghi các câu đã lọc vào các file mới
    write_conllu(output1_path, filtered_sentences_file1)
    write_conllu(output2_path, filtered_sentences_file2)

# Đường dẫn tới các file
file1_path = './content/output_final.txt'
file2_path = './content/output_final2.txt'
output1_path = './content/a_x.txt'
output2_path = './content/b_x.txt'

# Thực hiện lọc và ghi kết quả vào file mới
filter_sentences_by_common_roots(file1_path, file2_path, output1_path, output2_path)
```
